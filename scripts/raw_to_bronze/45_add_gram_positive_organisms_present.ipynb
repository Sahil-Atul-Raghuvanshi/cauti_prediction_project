```python
"""
Feature: gram_positive_organisms_present
Flag if gram-positive organisms found in urine culture
"""
import pandas as pd
import os
from config_paths import *
from utils import drop_columns
```

```python
# Read dataset
df = pd.read_csv(dataset_path)
initial_row_count = len(df)

# Load microbiology events
microbiology_events = pd.read_csv(
    os.path.join(hosp_path, "microbiologyevents.csv"),
    usecols=["subject_id", "hadm_id", "spec_type_desc", "test_name", "org_name"]
)

# Filter urine culture
urine_culture_df = microbiology_events[
    (microbiology_events["spec_type_desc"].str.upper() == "URINE") &
    (microbiology_events["test_name"].str.upper() == "URINE CULTURE")
].copy()

# Gram-positive organisms list (complete from notebook)
gram_positive_organisms = [
    "Staphylococcus, coagulase negative",
    "Staph aureus coag +",
    "Staphylococcus saprophyticus",
    "Staphylococcus lugdunensis",
    "Staphylococcus epidermidis",
    "Staphylococcus haemolyticus",
    "Staphylococcus species",
    "Enterococcus faecalis",
    "Enterococcus faecium",
    "Enterococcus sp.",
    "Probable Enterococcus",
    "Enterococcus hirae",
    "Streptococcus anginosus (Milleri) group",
    "Streptococcus species",
    "Beta streptococcus",
    "Beta streptococcus group A",
    "Beta streptococcus group B",
    "Beta streptococcus group G",
    "Presumptive Streptococcus bovis",
    "Viridans streptococci",
    "Corynebacterium species (diphtheroids)",
    "Corynebacterium urealyticum sp. nov.",
    "Lactobacillus species",
    "Micrococcus/Stomatococcus species",
    "Aerococcus viridans",
    "Aerococcus urinae",
    "Bacillus species; not anthracis",
    "Gram positive cocci",
    "Gram positive coccus(cocci)",
    "Gram positive rods",
    "Gram positive bacteria"
]

gram_positive_organisms_lower = [org.strip().lower() for org in gram_positive_organisms]

# Flag gram-positive (using .isin() for exact match like notebook)
urine_culture_df["is_gram_positive"] = (
    urine_culture_df["org_name"]
    .astype(str)
    .str.strip()
    .str.lower()
    .isin(gram_positive_organisms_lower)
)

# Admission-level flag (groupby hadm_id only, like notebook)
admission_gram_positive = (
    urine_culture_df
    .groupby("hadm_id")["is_gram_positive"]
    .any()
    .reset_index()
    .rename(columns={"is_gram_positive": "gram_positive_organisms_present"})
)

# Merge (on hadm_id only, like notebook)
df = df.merge(admission_gram_positive, on="hadm_id", how="left")

# Fill missing with False (using .where() like notebook)
df["gram_positive_organisms_present"] = df["gram_positive_organisms_present"].where(
    df["gram_positive_organisms_present"].notna(), False
).astype(bool)

# Save
df.to_csv(dataset_path, index=False)

# Diagnostic output
print(f"\nTotal rows: {len(df)}")
print(f"Expected total: {initial_row_count}")
if len(df) == initial_row_count:
    print(f"✓ Row count matches expected total ({initial_row_count})")
else:
    print(f"⚠ Row count mismatch! Expected {initial_row_count}, got {len(df)} (difference: {len(df) - initial_row_count})")
print(f"Dataset shape: {df.shape}")
print(f"Feature 'gram_positive_organisms_present' added. True count: {df['gram_positive_organisms_present'].sum()}")
print(f"Dataset shape: {df.shape}")

```
